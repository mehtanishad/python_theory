# What is Software?

Software is a set of instructions, data, or programs used to operate a computer and execute specific tasks. Without software, most computers would be useless. The majority of software is written in high-level programming languages due to the language being closer to natural human language as opposed to machine language. The high-level language is then translated into low-level machine code using a compiler or interpreter for the computer to understand. Software can also be written in a low-level assembly language, but it is less common.

# What is the purpose continue statement in python?
 
purpose continue statement
    the  continue keyword is used to end the current iteration in a for loop (or a while loop), and continues to the next iteration.
  

```python
# example:

for i in range(9):
  if i == 3:
    continue
  print(i)
```

# How will you compare two lists?

Comparison is the process when the data items of are checked against another data item of list, whether they are the same or not.


```python
# example
list1 = [11, 12, 13, 14, 15]  
list2 = [12, 13, 11, 15, 14]  
  
a = set(list1)  
b = set(list2)  
  
if a == b:  
    print("The list1 and list2 are equal")  
else:  
    print("The list1 and list2 are not equal")  
```

#  How will you create a dictionary using tuples in python?

convert a tuple to dictionary in Python, use the dict() method. The dict() function takes a tuple of tuples as an argument and returns the dictionary. Each tuple contains a key-value pair.


```python
# example
tup = ((11, "eleven"), (21, "mike"), (19, "dustin"), (46, "caleb"))
print(tup)

dct = dict((y, x) for x, y in tup)
print(dct)
```
